```python
def sum_of_first_n(n: int) -> int:
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
```

```python
sum_of_first_n(10)
```

```python
def gauss_sum(n: int) -> int:
    total = (n*(n + 1))/2 
    return int(total)
```

```python
gauss_sum(10)
```

```python
for i in range(10000):
    assert sum_of_first_n(i) == gauss_sum(i)
```

```python
%timeit sum_of_first_n(10000)
%timeit gauss_sum(10000)
```

```python
import time

def calculate_runtimes(func, n_arg, n_avg):
    runtimes = []
    for i in range(n_avg):
        t_0 = time.time()
        func(n_arg)
        t_1 = time.time()
        runtimes.append(t_1 - t_0)
    return sum(runtimes)/float(len(runtimes))
```

```python
calculate_runtimes(sum_of_first_n, 10000, 100)
```

```python
sum_of_first_n_times = []
gauss_sum_times = []
n_avg = 100
n_range = range(10, 100000, 2000)

for i in n_range:
    sum_of_first_n_times.append(calculate_runtimes(sum_of_first_n, i, n_avg))
    gauss_sum_times.append(calculate_runtimes(gauss_sum, i, n_avg))
```

```python
from matplotlib import pyplot as plt

plt.plot(n_range, sum_of_first_n_times, '.-', label = "sum_of_first_n_times")
plt.plot(n_range, gauss_sum_times, '.-', label = "gauss_sum_times")
plt.xlabel("n")
plt.ylabel("Ø Laufzeit (in ms)")
plt.legend();
```
