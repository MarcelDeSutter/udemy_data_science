```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
class ContinuousRandomVariable:
    def __init__(self, a = 0, b = 1):
        self.low = a
        self.high = b
        
    def draw(self, n_samples):
        samples = np.random.uniform(self.low, self.high, n_samples)
        return samples
```

```python
Zufallspunkt = ContinuousRandomVariable(0, 2)
Zufallspunkt.draw(10)
```

```python
sample = Zufallspunkt.draw(10000)
ax = sns.distplot(sample,
                  bins = 50,
                  kde = True,
                  color = "skyblue",
                  hist_kws = {"linewidth": 15,"alpha": 1})
ax.set(xlabel = "Ausprägungen", ylabel = "Häufigkeit", ylim = (0, 1));
```

```python
a = Zufallspunkt.low
b = Zufallspunkt.high
x = np.linspace(a - 2, b + 2, 1000)
y = np.array([1/(b-a) for i in x])
y[x < a] = 0
y[x > b] = 0

plt.plot(x, y)
plt.ylim(0, 1)
plt.xlabel("Wertebereich")
plt.ylabel("Wahrscheinlichkeit")
plt.legend(["Stetig Uniform"]);
```
